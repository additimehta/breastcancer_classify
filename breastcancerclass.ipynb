```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

##### Data Collection

```python
breast_cancer_ds = sklearn.datasets.load_breast_cancer()
df = pd.DataFrame(breast_cancer_ds.data, columns = breast_cancer_ds.feature_names )
```

```python
df.head()
```

```python
df['diagnosis'] = breast_cancer_ds.target
```

```python
df.shape
```

```python
df.info()
```

```python
# check for missing values
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
# let's check how many patients have benign or malign 
df['diagnosis'].value_counts()
#1 is for benign
#0 is for malign
```

```python
df.groupby('diagnosis').mean()
```

#### Seperating features and target

```python
X = df.drop(columns='diagnosis', axis=1)
Y = df['diagnosis']
```

```python
print(X)
```

```python
print(Y)
```

#### Splitting the data into training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#### Model Training using Logistic Regression

```python
model = LogisticRegression(max_iter=10000) 
model.fit(X_train, Y_train)
```

#### Model Evaluation

#### Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy on training data = ', training_data_accuracy)
```

```python
# accuracy on testing data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy on training data = ', test_data_accuracy)
```
